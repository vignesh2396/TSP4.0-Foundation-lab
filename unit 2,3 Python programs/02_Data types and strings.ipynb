# Finding the Body Mass Index

Find the BMI of the person using their height and weight. Formula - W/(h)2

[BMI](https://cdn.fs.teachablecdn.com/qTOp8afxSkGfU5YGYf36)

```python
weight_in_kg = int(input("Enter your Weight in Kg : "))
height_in_meter = float(input("Enter your height in meter"))
body_mass_index = round(weight_in_kg/(height_in_meter ** 2))
print(f"your BMI is :{body_mass_index}")
```
