```python
import matplotlib.pyplot as plt
```

# line plot

```python
x_axis = [1,2,3]
y_axis = [4,5,6]

# line plot
plt.plot(x_axis, y_axis)
```

```python
plt.plot(x_axis, y_axis)
plt.show()
```

```python
help(plt.plot)
```

# color

```python
plt.plot(x_axis, y_axis, color='g' )
plt.show()
```

# marker

```python
plt.plot(x_axis, y_axis, marker='x')
plt.show()
```

# title

```python
plt.title('Line Plot')
plt.plot(x_axis, y_axis, marker='x')
plt.show()
```

# label for x and y axis

```python
plt.title('Line Plot')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.plot(x_axis, y_axis, marker='x')
plt.show()
```

```python
x1_axis = [2,4,6,8]
y1_axis = [1,10,100,1000]
x2_axis = [1,3,5,7]
y2_axis = [100,110, 120,130]
plt.plot(x1_axis, y1_axis, 'r')
plt.plot(x2_axis, y2_axis, 'g')
plt.show()
```

# legend

```python
plt.plot(x1_axis, y1_axis, 'r')
plt.plot(x2_axis, y2_axis, 'g')

plt.legend(['Exponential values', 'linear values'])
plt.show()
```

# histograms

```python
values = [10, 15, 20, 10, 15]

plt.hist(values, width=0.5)
plt.show()
```

# bar plot

```python
values = [10, 15, 20, 30]
plt.bar(values, [0.5, 1, 2, 3], width=2)

plt.show()
```
