# matplotlib

```python
import matplotlib.pyplot as plt
```

## pie chart

```python
values = [20, 20, 35,25]
```

```python
plt.pie(values)
plt.show()
```

```python
plt.pie(values, labels = ['item1', 'item2', 'item3', 'item4'])
plt.show()
```

```python
plt.pie(values, labels = ['item1', 'item2', 'item3', 'item4'],
        explode=[0.0,0.0,0.2,0.0])
plt.show()
```

## scatter plot

```python
x_axis = list(range(1,11))
y_axis = [15,12,54,49,87,96,75,14,23,1]
```

```python
x_axis
```

```python
plt.title('Scatter Plot')
plt.xlabel('demand')
plt.ylabel('price')

plt.scatter(x_axis, y_axis)
plt.show()
```

```python
plt.title('Scatter Plot')
plt.xlabel('demand')
plt.ylabel('price')

plt.scatter(x_axis, y_axis, c=x_axis)
plt.show()
```

```python
plt.title('Scatter Plot')
plt.xlabel('demand')
plt.ylabel('price')

plt.scatter(x_axis, y_axis, c=x_axis, cmap='Accent')
plt.show()
```

## Handling dates

```python
dates = ['2022-01-01', '2021-01-01', '2020-01-01', '2019-01-01']
values = [100, 101, 102, 103]

plt.xticks(rotation=90)
plt.plot(dates, values)
plt.show()
```

## creating multiple subplots in one figure

```python
x1_axis = [2,4,6,8]
y1_axis = [1,10,100,1000]

x2_axis = [1,3,5,7]
y2_axis = [100,110, 120,130]
```

```python
# help(plt.subplot)
```

```python
#plt.subplot(no_of_rows,no_of_columns,current_plot_no)
plt.subplot(1,2,1)
plt.plot(x1_axis, y1_axis, 'r')

plt.subplot(1,2,2)
plt.plot(x2_axis, y2_axis, 'g')

plt.tight_layout()
```

```python
plt.subplot(2,1,1)
plt.plot(x1_axis, y1_axis, 'r')

plt.subplot(2,1,2)
plt.plot(x2_axis, y2_axis, 'g')

plt.tight_layout()
```

```python
plt.subplot(2,1,1)
plt.title('Exponential')
plt.plot(x1_axis, y1_axis, 'r')

plt.subplot(2,1,2)
plt.title('Linear')
plt.plot(x2_axis, y2_axis, 'g')

plt.tight_layout()
```
